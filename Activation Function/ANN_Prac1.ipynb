## **Importing Python Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **Defining Sigmoid Activation Function**

```python
def SigmoidBinary(t):
  return 1/(1 + np.exp(-t))
```

## **Plotting Sigmoid Function**

```python
t = np.linspace(-5, 5)
plt.title('Sigmoid Binary Function')
plt.plot(t, SigmoidBinary(t))
plt.scatter(t, SigmoidBinary(t))
plt.show
```

## **Defining Hyperbolic Tan Function**

```python
def HyperbolicTan(t):
  return (np.tanh(-t))
```

## **Plotting Hyperbolic Tan Funtion**

```python
t = np.linspace(-5, 5)
plt.title('HyperbolicTan Function')
plt.plot(t, HyperbolicTan(t))
plt.scatter(t, HyperbolicTan(t))
plt.show
```
